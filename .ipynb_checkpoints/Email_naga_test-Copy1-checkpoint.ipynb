```python
result, data = mail.uid('search', None, "ALL")
# search and return uids instead
i = len(data[0].split()) # data[0] is a space separate string
for x in range(i):
    latest_email_uid = data[0].split()[] # unique ids wrt label selected
    result, email_data = mail.uid('fetch', latest_email_uid, '(RFC822)')
 # fetch the email body (RFC822) for the given ID
    raw_email = email_data[0][1]
    #continue inside the same for loop as above
    raw_email_string = raw_email.decode('utf-8')
    # converts byte literal to string removing b''
    email_message = email.message_from_string(raw_email_string)
#     # this will loop through all the available multiparts in mail
    for part in email_message.walk():
        if part.get_content_type() == "text/html": # ignore attachments/html
            body = part.get_payload(decode=True)
#             print(body)
#             print(type(body))
            save_string = str("save" + str(x) + ".htm")
#             with io.open(join(getcwd(), 'Fwd Fly Morning News.htm'), 'r') as f:
#                     text = f.read()
#                     text.encode('utf-8')
    #unicodedata.normalize('NFKD', text).encode('ascii', 'ignore')
   
            # location on disk
            myfile = open(save_string, 'w')
#             print(type(myfile))
            myfile.write(body.decode("utf-8"))
#             # body is again a byte literal
            myfile.close()
        else:
            continue
```

```python
import pandas as pd
l=[]
print
for tick in soup.select('a[href^="https://thefly.com/news.php?symbol"]') :
        titles = tick.parent.parent.parent.select('a[href^="https://thefly.com/permalinks/entry.php"] span')
#         print(titles)
        if(not(',' in tick) and len(titles)>0):
            dict1 ={}
            dict1["Tick"]=tick.text 

            dict1["News"]=titles[0].text
    #         print(dict1)
            l.append(dict1)
#  
            
#        
df =pd.DataFrame(l)
df.head()           
```

```python
my_list=[]
for index, row in df.iterrows(): 
#     print (row["Tick"]) 
    mydict={}
    if not "," in row["Tick"] :
#         print(row["Tick"])
        mydict["TickerName"]=row["Tick"]
        mydict["TickerNews"]=row["News"]
        my_list.append(mydict)
flydf=pd.DataFrame(my_list)   
flydf.head()
```

```python
group_ticker=flydf.groupby("TickerName").count()
group_ticker=group_ticker.reset_index()
group_ticker.head()
```

```python
sort_ticker=group_ticker.sort_values("TickerNews",ascending=False)
sort_ticker.head()
```

```python
merged_df=pd.merge(flydf,sort_ticker,on=["TickerName"])
merged_df.head()
```

```python
final_tickerdf=merged_df.rename(columns={"TickerNews_x": "TickerNews", "TickerNews_y": "Count"})
# final_tickerdf
final_tickerdf=final_tickerdf.sort_values(["Count","TickerName"],ascending=False)
final_tickerdf.head()
# final_tickerdf.to_csv("flymorningnews.csv",index=False)
# y.to_csv("final_flymorning_news.csv")
```
